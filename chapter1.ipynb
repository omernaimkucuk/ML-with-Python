```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns   
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
df.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, iris.target, test_size=0.20, random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
combinations= []
for i in range(4):
    for j in range(i+1,4):
        combinations.append((i,j))
print(combinations)
fig, ax= plt.subplots(4, 4, figsize=(15,15))
for i in range(4):
    for j in range(4):
    
        if (i,j) in combinations:
            
            ax[i, j].scatter(X_train.iloc[:, i], X_train.iloc[:, j], c=y_train, cmap='viridis', alpha=0.7)
            ax[i, j].set_xlabel(X_train.columns[i])
            ax[i, j].set_ylabel(X_train.columns[j])

        else:
            ax[i, j].axis('off')


        
        
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn= KNeighborsClassifier(algorithm='auto', leaf_size=30, metric='minkowski', metric_params=None, n_jobs=1, n_neighbors=1, p=2, weights='uniform')
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
np.mean(y_pred == y_test)
```

```python
knn.score(X_test, y_test)
```
